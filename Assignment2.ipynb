```python
import pandas as pd

# Sample data
data = {
    'prompt': [
        "Create a chatbot script for customer support",
        "Generate a response for mental health chat",
        "Write a conversation starter for virtual assistants",
        "Generate an image of a sunset over the ocean",
        "Create a video clip of a relaxing forest scene"
    ],
    'category': [
        "Communication",
        "Communication",
        "Communication",
        "Visual Art",
        "Visual Art"
    ],
    'subcategory': [
        "Chatbots and Virtual Assistants",
        "Mental health",
        "Chatbots and Virtual Assistants",
        "Image Generation",
        "Video Generation"
    ]
}

df = pd.DataFrame(data)


```

```python
# Define rule-based function
def classify_prompt(prompt):
    if 'chatbot' in prompt or 'virtual assistant' in prompt:
        return "Communication", "Chatbots and Virtual Assistants"
    elif 'mental health' in prompt:
        return "Communication", "Mental health"
    elif 'image' in prompt:
        return "Visual Art", "Image Generation"
    elif 'video' in prompt:
        return "Visual Art", "Video Generation"
    else:
        return "Unknown", "Unknown"
```

```python

# Test the function
test_prompt = "Create a chatbot script for customer support"
category, subcategory = classify_prompt(test_prompt)
print(f"Category: {category}, Subcategory: {subcategory}")
```
